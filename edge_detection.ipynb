#### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from matplotlib import image
import scipy.stats as stats
```

#### Creating Gaussian filter

```python
fx=stats.norm.pdf(np.linspace(-6,6,50),loc=0, scale=2)
filt=np.outer(fx,fx)
plt.imshow(filt, cmap="gray");
```

#### Grayscale conversion and filtered image

```python
im=image.imread("dogs.jpeg")
im_arr=np.asarray(im)
gray=np.mean(im_arr,axis=2)
out=convolve2d(gray,filt)
grayim=np.dot(im_arr[:,:,:3],[0.2989, 0.5870, 0.1140])
fig,ax=plt.subplots(1,4, figsize=(17,8))
ax[0].imshow(im_arr)
ax[2].imshow(gray, cmap="gray")
ax[3].imshow(out, cmap="gray")
ax[1].imshow(grayim, cmap="gray");
```

#### Edge Detection

```python
hx=np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
hy=np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
gx1=convolve2d(hx,grayim)
gy1=convolve2d(hy,grayim)
gx2=convolve2d(hx,gray)
gy2=convolve2d(hy,gray)
```

```python
g1=np.sqrt(np.multiply(gx1,gx1)+ np.multiply(gy1,gy1))
g2=np.sqrt(np.multiply(gx2,gx2)+ np.multiply(gy2,gy2))
```

```python
fig,ax=plt.subplots(1,2, figsize=(17,4))
ax[0].imshow(g1, cmap="gray")
ax[1].imshow(g2, cmap="gray");
```
